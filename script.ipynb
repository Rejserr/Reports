# [dbo].[v_pickingStorageSystem]

```sql
/****** Object:  View [dbo].[v_pickingStorageSystem]    Script Date: 16.04.2025. 14:01:36 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Kreiranje view-a v_pickingStorageSystem
CREATE VIEW [dbo].[v_pickingStorageSystem]
AS
WITH PickingData AS (
    SELECT 
        CAST(FromLocationCode AS NVARCHAR(50)) COLLATE DATABASE_DEFAULT AS normalized_location_code,
        ID,
        OrderCode,
        OrderInputDate,
        DeliveryType,
        TaskCode,
        ListCode,
        TaskCreateTime,
        PickDateTime,
        LogID,
        FromLocationCode,
        FromSSCC,
        ToSSCC,
        ItemCode,
        ItemName,
        Qty,
        UserName,
        Customer,
        Receiver,
        DeliveryAddress,
        City,
        ZipCode,
        ImportDate,
        StartDateParam,
        EndDateParam
    FROM 
        [Reports].[dbo].[Picking]
)
SELECT 
    -- Sva polja iz tablice Picking
    p.ID,
    p.OrderCode,
    p.OrderInputDate,
    p.DeliveryType,
    p.TaskCode,
    p.ListCode,
    p.TaskCreateTime,
    p.PickDateTime,
    p.LogID,
    p.FromLocationCode,
    
    -- Dodana kolona Storage_system iz v_SystemStorage ili iz Locations tablice
    COALESCE(
        s.Storage_system, 
        (SELECT TOP 1 Storage_system FROM [Reports].[dbo].[Locations] 
         WHERE Code COLLATE DATABASE_DEFAULT = p.normalized_location_code)
    ) AS Storage_system,
    
    p.FromSSCC,
    p.ToSSCC,
    p.ItemCode,
    p.ItemName,
    p.Qty,
    p.UserName,
    p.Customer,
    p.Receiver,
    p.DeliveryAddress,
    p.City,
    p.ZipCode,
    p.ImportDate,
    p.StartDateParam,
    p.EndDateParam
FROM 
    PickingData p
LEFT JOIN 
    (
        SELECT 
            CAST(Code AS NVARCHAR(50)) COLLATE DATABASE_DEFAULT AS normalized_code,
            Code,
            Storage_system
        FROM 
            [Reports].[dbo].[v_SystemStorage]
        GROUP BY
            Code, Storage_system
    ) s ON p.normalized_location_code = s.normalized_code
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[sp_Import_Picking]    Script Date: 16.04.2025. 14:01:36 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Kreirajmo novu proceduru koja će sigurno raditi
CREATE PROCEDURE [dbo].[sp_Import_Picking]
    @StartDate VARCHAR(50),
    @EndDate VARCHAR(50),
    @DepositorCode VARCHAR(50) = '*',
    @LocationCode VARCHAR(50) = '*',
    @StatusID INT = -1,
    @UserID INT = -1,
    @TransactionTypeID INT = -1,
    @ProductCode VARCHAR(50) = '*',
    @OrderTypeID INT = -1,
    @CustomerID INT = -1
AS
BEGIN
    SET NOCOUNT ON;
    
    -- Prvo provjerimo postoji li problematični LogID u tablici
    IF EXISTS (SELECT 1 FROM [dbo].[Picking] WHERE LogID = 423300)
    BEGIN
        PRINT 'Problematični LogID 423300 već postoji u tablici Picking. Preskačemo ga.';
    END
    
    -- Kreirajmo tablicu za pohranu rezultata
    DECLARE @InsertedRows INT = 0;
    DECLARE @TotalRows INT = 0;
    
    -- Pripremimo SQL za dohvat podataka iz funkcije
    DECLARE @RemoteSQL NVARCHAR(MAX);
    SET @RemoteSQL = 'SELECT 
        OrderCode,
        OrderInputDate,
        DeliveryType,
        TaskCode,
        ListCode,
        TaskCreateTime,
        PickDateTime,
        LogID,
        FromLocationCode,
        FromSSCC,
        ToSSCC,
        ItemCode,
        ItemName,
        Qty,
        UserName,
        Customer,
        Receiver,
        DeliveryAddress,
        City,
        ZipCode
    FROM LVision.dbo.fn_RPT_Picking(''' 
                   + @StartDate + ''', ''' 
                   + @EndDate + ''', ''' 
                   + @DepositorCode + ''', ''' 
                   + @LocationCode + ''', ' 
                   + CAST(@StatusID AS VARCHAR) + ', ' 
                   + CAST(@UserID AS VARCHAR) + ', ' 
                   + CAST(@TransactionTypeID AS VARCHAR) + ', ''' 
                   + @ProductCode + ''', ' 
                   + CAST(@OrderTypeID AS VARCHAR) + ', ' 
                   + CAST(@CustomerID AS VARCHAR) + ')';
    
    -- Kreirajmo privremenu tablicu za pohranu rezultata
    CREATE TABLE #SourceData (
        OrderCode NVARCHAR(50),
        OrderInputDate DATETIME,
        DeliveryType NVARCHAR(50),
        TaskCode NVARCHAR(50),
        ListCode NVARCHAR(50),
        TaskCreateTime DATETIME,
        PickDateTime DATETIME,
        LogID INT,
        FromLocationCode NVARCHAR(50),
        FromSSCC NVARCHAR(50),
        ToSSCC NVARCHAR(50),
        ItemCode NVARCHAR(50),
        ItemName NVARCHAR(500),
        Qty DECIMAL(18,6),
        UserName NVARCHAR(100),
        Customer NVARCHAR(200),
        Receiver NVARCHAR(200),
        DeliveryAddress NVARCHAR(200),
        City NVARCHAR(100),
        ZipCode NVARCHAR(50)
    );
    
    -- Dohvatimo podatke iz funkcije
    DECLARE @SQL NVARCHAR(MAX);
    SET @SQL = 'INSERT INTO #SourceData
    SELECT DISTINCT * FROM OPENQUERY(LVS, ''' + REPLACE(@RemoteSQL, '''', '''''') + ''')';
    
    BEGIN TRY
        -- Izvršimo SQL za dohvat podataka
        EXEC sp_executesql @SQL;
        
        -- Dohvatimo ukupan broj redova
        SELECT @TotalRows = COUNT(*) FROM #SourceData;
        PRINT 'Ukupno dohvaćeno redova: ' + CAST(@TotalRows AS VARCHAR);
        
        -- Provjerimo ima li duplikata
        DECLARE @DuplicateCount INT;
        SELECT @DuplicateCount = COUNT(*) - COUNT(DISTINCT LogID) FROM #SourceData;
        
        IF @DuplicateCount > 0
        BEGIN
            PRINT 'Upozorenje: Pronađeno ' + CAST(@DuplicateCount AS VARCHAR) + ' duplikata u izvornim podacima!';
            
            -- Uklonimo duplikate
            DELETE s1
            FROM #SourceData s1
            INNER JOIN (
                SELECT LogID, MIN(OrderCode) AS MinOrderCode
                FROM #SourceData
                GROUP BY LogID
                HAVING COUNT(*) > 1
            ) s2 ON s1.LogID = s2.LogID AND s1.OrderCode <> s2.MinOrderCode;
            
            -- Provjerimo broj redova nakon uklanjanja duplikata
            SELECT @TotalRows = COUNT(*) FROM #SourceData;
            PRINT 'Broj redova nakon uklanjanja duplikata: ' + CAST(@TotalRows AS VARCHAR);
        END
        
        -- Umetnimo podatke u tablicu Picking, jedan po jedan
        DECLARE @CurrentLogID INT;
        DECLARE @CurrentOrderCode NVARCHAR(50);
        DECLARE @CurrentOrderInputDate DATETIME;
        DECLARE @CurrentDeliveryType NVARCHAR(50);
        DECLARE @CurrentTaskCode NVARCHAR(50);
        DECLARE @CurrentListCode NVARCHAR(50);
        DECLARE @CurrentTaskCreateTime DATETIME;
        DECLARE @CurrentPickDateTime DATETIME;
        DECLARE @CurrentFromLocationCode NVARCHAR(50);
        DECLARE @CurrentFromSSCC NVARCHAR(50);
        DECLARE @CurrentToSSCC NVARCHAR(50);
        DECLARE @CurrentItemCode NVARCHAR(50);
        DECLARE @CurrentItemName NVARCHAR(500);
        DECLARE @CurrentQty DECIMAL(18,6);
        DECLARE @CurrentUserName NVARCHAR(100);
        DECLARE @CurrentCustomer NVARCHAR(200);
        DECLARE @CurrentReceiver NVARCHAR(200);
        DECLARE @CurrentDeliveryAddress NVARCHAR(200);
        DECLARE @CurrentCity NVARCHAR(100);
        DECLARE @CurrentZipCode NVARCHAR(50);
        
        -- Kreirajmo kursor za prolazak kroz podatke
        DECLARE data_cursor CURSOR FOR
        SELECT 
            OrderCode,
            OrderInputDate,
            DeliveryType,
            TaskCode,
            ListCode,
            TaskCreateTime,
            PickDateTime,
            LogID,
            FromLocationCode,
            FromSSCC,
            ToSSCC,
            ItemCode,
            ItemName,
            Qty,
            UserName,
            Customer,
            Receiver,
            DeliveryAddress,
            City,
            ZipCode
        FROM #SourceData;
        
        OPEN data_cursor;
        
        FETCH NEXT FROM data_cursor INTO
            @CurrentOrderCode,
            @CurrentOrderInputDate,
            @CurrentDeliveryType,
            @CurrentTaskCode,
            @CurrentListCode,
            @CurrentTaskCreateTime,
            @CurrentPickDateTime,
            @CurrentLogID,
            @CurrentFromLocationCode,
            @CurrentFromSSCC,
            @CurrentToSSCC,
            @CurrentItemCode,
            @CurrentItemName,
            @CurrentQty,
            @CurrentUserName,
            @CurrentCustomer,
            @CurrentReceiver,
            @CurrentDeliveryAddress,
            @CurrentCity,
            @CurrentZipCode;
        
        WHILE @@FETCH_STATUS = 0
        BEGIN
            -- Provjerimo postoji li već zapis s istim LogID
            IF NOT EXISTS (SELECT 1 FROM [dbo].[Picking] WHERE LogID = @CurrentLogID)
            BEGIN
                -- Umetnimo novi zapis
                BEGIN TRY
                    INSERT INTO [dbo].[Picking]
                    (
                        OrderCode,
                        OrderInputDate,
                        DeliveryType,
                        TaskCode,
                        ListCode,
                        TaskCreateTime,
                        PickDateTime,
                        LogID,
                        FromLocationCode,
                        FromSSCC,
                        ToSSCC,
                        ItemCode,
                        ItemName,
                        Qty,
                        UserName,
                        Customer,
                        Receiver,
                        DeliveryAddress,
                        City,
                        ZipCode,
                        ImportDate,
                        StartDateParam,
                        EndDateParam
                    )
                    VALUES
                    (
                        @CurrentOrderCode,
                        @CurrentOrderInputDate,
                        @CurrentDeliveryType,
                        @CurrentTaskCode,
                        @CurrentListCode,
                        @CurrentTaskCreateTime,
                        @CurrentPickDateTime,
                        @CurrentLogID,
                        @CurrentFromLocationCode,
                        @CurrentFromSSCC,
                        @CurrentToSSCC,
                        @CurrentItemCode,
                        @CurrentItemName,
                        @CurrentQty,
                        @CurrentUserName,
                        @CurrentCustomer,
                        @CurrentReceiver,
                        @CurrentDeliveryAddress,
                        @CurrentCity,
                        @CurrentZipCode,
                        GETDATE(),
                        @StartDate,
                        @EndDate
                    );
                    
                    SET @InsertedRows = @InsertedRows + 1;
                END TRY
                BEGIN CATCH
                    -- Ako dođe do greške, samo preskočimo taj zapis i nastavimo dalje
                    PRINT 'Preskačem LogID: ' + CAST(@CurrentLogID AS VARCHAR) + ' zbog greške: ' + ERROR_MESSAGE();
                END CATCH
            END
            
            FETCH NEXT FROM data_cursor INTO
                @CurrentOrderCode,
                @CurrentOrderInputDate,
                @CurrentDeliveryType,
                @CurrentTaskCode,
                @CurrentListCode,
                @CurrentTaskCreateTime,
                @CurrentPickDateTime,
                @CurrentLogID,
                @CurrentFromLocationCode,
                @CurrentFromSSCC,
                @CurrentToSSCC,
                @CurrentItemCode,
                @CurrentItemName,
                @CurrentQty,
                @CurrentUserName,
                @CurrentCustomer,
                @CurrentReceiver,
                @CurrentDeliveryAddress,
                @CurrentCity,
                @CurrentZipCode;
        END
        
        CLOSE data_cursor;
        DEALLOCATE data_cursor;
        
        PRINT 'Broj novih unesenih redova: ' + CAST(@InsertedRows AS VARCHAR);
        
        -- Vratimo rezultate
        SELECT @InsertedRows AS NewRecordsAdded, 
               (SELECT COUNT(*) FROM [dbo].[Picking]) AS TotalRecords;
    END TRY
    BEGIN CATCH
        -- Ako je kursor otvoren, zatvorimo ga
        IF (SELECT CURSOR_STATUS('global', 'data_cursor')) >= 0
        BEGIN
            CLOSE data_cursor;
            DEALLOCATE data_cursor;
        END
        
        -- Dohvatite informacije o grešci
        DECLARE @ErrorMessage NVARCHAR(4000) = ERROR_MESSAGE();
        DECLARE @ErrorSeverity INT = ERROR_SEVERITY();
        DECLARE @ErrorState INT = ERROR_STATE();
        
        -- Ispiši informacije o grešci
        PRINT 'Error: ' + @ErrorMessage;
        
        -- Ponovno bacite grešku koristeći RAISERROR
        RAISERROR(@ErrorMessage, @ErrorSeverity, @ErrorState);
    END CATCH
    
    -- Očistimo privremenu tablicu
    IF OBJECT_ID('tempdb..#SourceData') IS NOT NULL
        DROP TABLE #SourceData;
END
GO
```

# [dbo].[sp_View_Picking]

```sql
/****** Object:  StoredProcedure [dbo].[sp_View_Picking]    Script Date: 16.04.2025. 14:01:36 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Kreirajmo proceduru za pregled podataka
CREATE   PROCEDURE [dbo].[sp_View_Picking]
    @StartDate DATETIME = NULL,
    @EndDate DATETIME = NULL,
    @OrderCode NVARCHAR(50) = NULL,
    @ItemCode NVARCHAR(50) = NULL,
    @Customer NVARCHAR(200) = NULL,
    @City NVARCHAR(100) = NULL
AS
BEGIN
    SET NOCOUNT ON;
    
    SELECT 
        OrderCode,
        OrderInputDate,
        DeliveryType,
        TaskCode,
        ListCode,
        TaskCreateTime,
        PickDateTime,
        LogID,
        FromLocationCode,
        FromSSCC,
        ToSSCC,
        ItemCode,
        ItemName,
        Qty,
        UserName,
        Customer,
        Receiver,
        DeliveryAddress,
        City,
        ZipCode,
        ImportDate
    FROM 
        [dbo].[Picking]
    WHERE 
        (@StartDate IS NULL OR TaskCreateTime >= @StartDate) AND
        (@EndDate IS NULL OR TaskCreateTime <= @EndDate) AND
        (@OrderCode IS NULL OR OrderCode LIKE '%' + @OrderCode + '%') AND
        (@ItemCode IS NULL OR ItemCode LIKE '%' + @ItemCode + '%') AND
        (@Customer IS NULL OR Customer LIKE '%' + @Customer + '%') AND
        (@City IS NULL OR City LIKE '%' + @City + '%')
    ORDER BY 
        TaskCreateTime DESC
END
GO
```

# [dbo].[stock_details]

```sql
/****** Object:  StoredProcedure [dbo].[stock_details]    Script Date: 16.04.2025. 14:01:36 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE   PROCEDURE [dbo].[stock_details]
    @Depositors VARCHAR(500) = NULL, -- Comma-separated list of depositors (e.g., 'DEP1,DEP2')
    @Warehouses VARCHAR(500) = NULL, -- Comma-separated list of warehouses (e.g., 'WH1,WH2')
    @ItemCodes VARCHAR(500) = NULL   -- Comma-separated list of item codes (e.g., 'ITEM1,ITEM2')
AS
BEGIN
    SET NOCOUNT ON;
    
    -- If no filters specified, show all stock
    IF @Depositors IS NULL AND @Warehouses IS NULL AND @ItemCodes IS NULL
    BEGIN
        DECLARE @SQL1 NVARCHAR(MAX);
        
        SET @SQL1 = 'SELECT * FROM OPENQUERY(LVS, 
        ''SELECT *
        FROM [LVision].[dbo].[V_Stock](NOLOCK)
        WHERE LanguageID = ''''1''''
          AND LanguageID1 = ''''1''''
          AND LanguageID2 = 31
        ORDER BY ItemCode, WarehouseCode, LocationCode'')';
        
        EXEC sp_executesql @SQL1;
    END
    ELSE
    BEGIN
        -- Create a table to store individual filter queries
        CREATE TABLE #FilterQueries (QueryText NVARCHAR(MAX));
        
        -- Base query
        DECLARE @BaseQuery NVARCHAR(MAX) = 'SELECT * FROM OPENQUERY(LVS, 
        ''SELECT *
        FROM [LVision].[dbo].[V_Stock](NOLOCK)
        WHERE LanguageID = ''''1''''
          AND LanguageID1 = ''''1''''
          AND LanguageID2 = 31';
        
        -- Add depositor filter if provided
        IF @Depositors IS NOT NULL
        BEGIN
            DECLARE @DepositorsFilter NVARCHAR(MAX) = '';
            DECLARE @Depositor VARCHAR(50);
            DECLARE @Pos INT;
            
            SET @Depositors = @Depositors + ',';
            SET @Pos = CHARINDEX(',', @Depositors, 1);
            
            WHILE @Pos > 0
            BEGIN
                SET @Depositor = LTRIM(RTRIM(LEFT(@Depositors, @Pos - 1)));
                IF @Depositor <> ''
                BEGIN
                    IF @DepositorsFilter = ''
                        SET @DepositorsFilter = ' AND (DepositorCode = '''''+ @Depositor + '''''';
                    ELSE
                        SET @DepositorsFilter = @DepositorsFilter + ' OR DepositorCode = '''''+ @Depositor + '''''';
                END
                
                SET @Depositors = RIGHT(@Depositors, LEN(@Depositors) - @Pos);
                SET @Pos = CHARINDEX(',', @Depositors, 1);
            END
            
            IF @DepositorsFilter <> ''
                SET @DepositorsFilter = @DepositorsFilter + ')';
                
            SET @BaseQuery = @BaseQuery + @DepositorsFilter;
        END
        
        -- Add warehouse filter if provided
        IF @Warehouses IS NOT NULL
        BEGIN
            DECLARE @WarehousesFilter NVARCHAR(MAX) = '';
            DECLARE @Warehouse VARCHAR(50);
            
            SET @Warehouses = @Warehouses + ',';
            SET @Pos = CHARINDEX(',', @Warehouses, 1);
            
            WHILE @Pos > 0
            BEGIN
                SET @Warehouse = LTRIM(RTRIM(LEFT(@Warehouses, @Pos - 1)));
                IF @Warehouse <> ''
                BEGIN
                    IF @WarehousesFilter = ''
                        SET @WarehousesFilter = ' AND (WarehouseCode = '''''+ @Warehouse + '''''';
                    ELSE
                        SET @WarehousesFilter = @WarehousesFilter + ' OR WarehouseCode = '''''+ @Warehouse + '''''';
                END
                
                SET @Warehouses = RIGHT(@Warehouses, LEN(@Warehouses) - @Pos);
                SET @Pos = CHARINDEX(',', @Warehouses, 1);
            END
            
            IF @WarehousesFilter <> ''
                SET @WarehousesFilter = @WarehousesFilter + ')';
                
            SET @BaseQuery = @BaseQuery + @WarehousesFilter;
        END
        
        -- Add item code filter if provided
        IF @ItemCodes IS NOT NULL
        BEGIN
            DECLARE @ItemCodesFilter NVARCHAR(MAX) = '';
            DECLARE @ItemCode VARCHAR(50);
            
            SET @ItemCodes = @ItemCodes + ',';
            SET @Pos = CHARINDEX(',', @ItemCodes, 1);
            
            WHILE @Pos > 0
            BEGIN
                SET @ItemCode = LTRIM(RTRIM(LEFT(@ItemCodes, @Pos - 1)));
                IF @ItemCode <> ''
                BEGIN
                    IF @ItemCodesFilter = ''
                        SET @ItemCodesFilter = ' AND (ItemCode = '''''+ @ItemCode + '''''';
                    ELSE
                        SET @ItemCodesFilter = @ItemCodesFilter + ' OR ItemCode = '''''+ @ItemCode + '''''';
                END
                
                SET @ItemCodes = RIGHT(@ItemCodes, LEN(@ItemCodes) - @Pos);
                SET @Pos = CHARINDEX(',', @ItemCodes, 1);
            END
            
            IF @ItemCodesFilter <> ''
                SET @ItemCodesFilter = @ItemCodesFilter + ')';
                
            SET @BaseQuery = @BaseQuery + @ItemCodesFilter;
        END
        
        -- Complete the query
        SET @BaseQuery = @BaseQuery + ' ORDER BY ItemCode, WarehouseCode, LocationCode'')';
        
        -- Execute the query
        EXEC sp_executesql @BaseQuery;
    END
END
GO
```

# [dbo].[task_details]

```sql
/****** Object:  StoredProcedure [dbo].[task_details]    Script Date: 16.04.2025. 14:01:36 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE   PROCEDURE [dbo].[task_details]
    @Months VARCHAR(50) = NULL, -- Comma-separated list of months (e.g., '3,4')
    @Year INT = NULL
AS
BEGIN
    SET NOCOUNT ON;
    
    -- If no months specified, show all months
    IF @Months IS NULL
    BEGIN
        DECLARE @SQL1 NVARCHAR(MAX);
        
        SET @SQL1 = 'SELECT * FROM OPENQUERY(LVS, 
        ''SELECT * 
        FROM [LVISION].[dbo].[V_TaskDetails](NOLOCK)
        WHERE LanguageID = ''''1'''' 
          AND LanguageID2 = ''''1'''' 
          AND LanguageID3 = ''''1'''' 
          AND LanguageID5 = ''''1'''''')';
        
        -- Add year filter if provided
        IF @Year IS NOT NULL
        BEGIN
            SET @SQL1 = @SQL1 + ' WHERE YEAR(ses_StartTime) = ' + CAST(@Year AS NVARCHAR(4));
        END
        
        SET @SQL1 = @SQL1 + ' ORDER BY tsk_ID DESC';
        
        EXEC sp_executesql @SQL1;
    END
    ELSE
    BEGIN
        -- Create a table to store individual month queries
        CREATE TABLE #MonthQueries (QueryText NVARCHAR(MAX));
        
        -- Parse the comma-separated months
        DECLARE @Month VARCHAR(10)
        DECLARE @Pos INT
        
        SET @Months = @Months + ','
        SET @Pos = CHARINDEX(',', @Months, 1)
        
        WHILE @Pos > 0
        BEGIN
            SET @Month = LTRIM(RTRIM(LEFT(@Months, @Pos - 1)))
            IF @Month <> ''
            BEGIN
                -- Build individual query for each month
                DECLARE @MonthQuery NVARCHAR(MAX);
                SET @MonthQuery = 'SELECT * FROM OPENQUERY(LVS, 
                ''SELECT * 
                FROM [LVISION].[dbo].[V_TaskDetails](NOLOCK)
                WHERE LanguageID = ''''1'''' 
                  AND LanguageID2 = ''''1'''' 
                  AND LanguageID3 = ''''1'''' 
                  AND LanguageID5 = ''''1''''
                  AND MONTH(ses_StartTime) = ' + @Month;
                
                -- Add year filter if provided
                IF @Year IS NOT NULL
                BEGIN
                    SET @MonthQuery = @MonthQuery + ' AND YEAR(ses_StartTime) = ' + CAST(@Year AS NVARCHAR(4));
                END
                
                SET @MonthQuery = @MonthQuery + ''')';
                
                INSERT INTO #MonthQueries (QueryText) VALUES (@MonthQuery);
            END
            SET @Months = RIGHT(@Months, LEN(@Months) - @Pos)
            SET @Pos = CHARINDEX(',', @Months, 1)
        END
        
        -- Combine all month queries with UNION ALL
        DECLARE @FinalQuery NVARCHAR(MAX) = '';
        
        SELECT @FinalQuery = 
            CASE 
                WHEN @FinalQuery = '' THEN QueryText
                ELSE @FinalQuery + ' UNION ALL ' + QueryText
            END
        FROM #MonthQueries;
        
        -- Add ORDER BY at the end
        SET @FinalQuery = @FinalQuery + ' ORDER BY tsk_ID DESC';
        
        -- Execute the combined query
        EXEC sp_executesql @FinalQuery;
        
        -- Clean up
        DROP TABLE #MonthQueries;
    END
END
GO
```
